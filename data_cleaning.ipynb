```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
citibike = pd.read_csv("202109-citibike-tripdata.csv")
```

```python
citibike.head()
```

```python
# re-assign station id as integers
station_list = list(set(list(citibike["start_station_name"].dropna().unique()) + list(citibike["end_station_name"].dropna().unique())))
station_dict = {k: v for (v, k) in enumerate(sorted(station_list))}
```

```python
# day, time of day, and distance of each trip
start_day = []
start_hour = []
end_day = []
end_hour = []
time_list = []
distance_list = []

i = 0
for (x, y) in citibike.iterrows():
# convert timestamps to numerical values
    start_time = datetime.strptime(y["started_at"], "%Y-%m-%d %H:%M:%S")
    end_time = datetime.strptime(y["ended_at"], "%Y-%m-%d %H:%M:%S")
# hour value of start and end times    
    start_hour.append(start_time.hour)    
    end_hour.append(end_time.hour)
# weekday = 1, weekend = 0
    s_day = start_time.weekday()
    if s_day < 5:
        start_day.append(1)
    else:
        start_day.append(0)
    e_day = end_time.weekday()
    if e_day < 5:
        end_day.append(1)
    else:
        end_day.append(0)
# time from start to finish
    total_time = (end_time - start_time).total_seconds()
    time_list.append(total_time)
# distance between stations    
    lat = ((y["end_lat"] - y["start_lat"]) * 69) ** 2
    lng = ((y["end_lng"] - y["start_lng"]) * 54.6) ** 2
    dist = round((lat + lng) ** 0.5, 2)
    distance_list.append(dist)
    
    i += 1
    if i % 100000 == 0:
        print(i)
```

```python
# replace staton names with index numbers
start_list = []
end_list = []

j = 0
for (x, y) in citibike[["start_station_name", "end_station_name"]].fillna(-1).iterrows():
    if y["start_station_name"] != -1:
        start_list.append(station_dict[y["start_station_name"]])
    else:
        start_list.append(-1)
        
    if y["end_station_name"] != -1:
        end_list.append(station_dict[y["end_station_name"]])
    else:
        end_list.append(-1)
        
    j += 1
    if j % 100000 == 0:
        print(j)
```

```python
# change ride and membership types to numbers
ride_type = []
member = []
k = 0

for (x, y) in citibike[["rideable_type", "member_casual"]].iterrows():
    if y["rideable_type"] == "classic_bike":
        ride_type.append(0)
    elif y["rideable_type"] == "docked_bike":
        ride_type.append(1)
    else:
        ride_type.append(2)
    
    if y["member_casual"] == "member":
        member.append(0)
    else:
        member.append(1)
    
    k += 1
    if k % 100000 == 0:
        print(k)
```

```python
df = pd.DataFrame({
    "start_id": start_list,
    "start_day": start_day,
    "start_hour": start_hour,
    "end_id": end_list,
    "end_day": end_day,
    "end_hour": end_hour,
    "total_time": time_list,
    "total_distance": distance_list,
    "rideable_type": ride_type,
    "member_casual": member
})
```

```python
df.head()
```

```python
# remove trips lasting less than one minute or over one week
df = df.loc[df["total_time"] > 60].loc[df["total_time"] < 86400]
df
```

```python
df.to_csv("cleaned.csv", index=False)
```
